## You must manually seperate the test and train stuff. And put it in two seperate CSVs

```python
import os
import shutil
import openpyxl


excelSheet = '/home/hooman/Downloads/Boulder_after_May22.xlsx'

img_dest_dir = "/home/hooman/Downloads/boulderNew/images"

label_dest_dir = "/home/hooman/Downloads/boulderNew/labels"
```

```python
# get the excel workbook and the specific sheet (i.e. tab) in the workbook
excel_workBook = openpyxl.load_workbook(excelSheet)
excel_sheet = excel_workBook[excel_workBook.sheetnames[0]]
```

```python
excel_sheet
```

```python
# find the column number for scene images
for excel_col in range(excel_sheet.max_column):
    if excel_sheet[1][excel_col].value == "Custom field (Boulder - Image Directory)":
        imagesColumnNb = excel_col
    if excel_sheet[1][excel_col].value == "Custom field (Scene - Label Directory)":
        labelsColumnNb = excel_col
```

```python
# Print the rows of the specified column. Note that sceneAddresses[0] takes column name.
imageAddresses = []
labelAddresses = []
for row in range(excel_sheet.max_row):
    #get the row value and convert it to string
    tempStr = str(excel_sheet[row+1][imagesColumnNb].value)
    
    
    #Note that we could just use below if both where in linux but we need the os.normalPath to convert our / to \. 
    #***also note that os.path adds two backsplashes. but if you do print(somePath) you see only one.
    #tempStr = tempStr.replace("file:////motionmetrics.net/nas/", "N:\\")
    tempStr = tempStr.replace("file:////motionmetrics.net/nas/randd", "/home/hooman/randd")
    
    imageAddresses.append(os.path.normpath(tempStr))
    
    
    #do the same but with labelAddresses
    adStr = str(excel_sheet[row+1][labelsColumnNb].value)
    
    # in case you have a PNG directory
    adStr = adStr.replace("Image", "Label")
    adStr = adStr + "PNG"
    
    labelAddresses.append(os.path.normpath(adStr.replace("file:////motionmetrics.net/nas/randd", "/home/hooman/randd")))
```

```python
import os
# copy over images
for i in range(len(imageAddresses)-1):
    #gotta skip the first entry which is column name
    srcDir = imageAddresses[i+1]

    if os.path.isdir(srcDir):
        file_names = os.listdir(srcDir)
        print(file_names)
        for file_name in file_names:
            print(file_name)
            if file_name and '.png' in file_name:
            #if file_name and '.jpg' in file_name:
                full_file_name = os.path.join(srcDir, file_name)
                if (os.path.isfile(full_file_name)):
                    shutil.copy(full_file_name, img_dest_dir)
```

```python
# copy over labels
for i in range(len(labelAddresses)-1):
    #gotta skip the first entry which is column name
    srcDir = labelAddresses[i+1]
    
    if os.path.isdir(srcDir):
        file_names = os.listdir(srcDir)
        print(file_names)
        for file_name in file_names:
            if '.png' in file_name:
                full_file_name = os.path.join(srcDir, file_name)
                if (os.path.isfile(full_file_name)):
                    shutil.copy(full_file_name, label_dest_dir)
```

# Remove the images for which we have no labels

```python
labelFileNameDic = {}

for file in os.listdir(label_dest_dir):
    fileName = file.replace(".png", "")
    if fileName in labelFileNameDic:
        labelFileNameDic[fileName] += 1
    else:
        labelFileNameDic[fileName] = 0
```

```python
imageFileNameDic = {}

for file in os.listdir(img_dest_dir):
    #fileName = file.replace(".jpg", "")
    fileName = file.replace(".png", "")
    if fileName in imageFileNameDic:
        imageFileNameDic[fileName] += 1
    else:
        imageFileNameDic[fileName] = 0
```

```python
print(len(imageFileNameDic))
print(len(labelFileNameDic))
```

```python
# remove images we do not have labels for

for name in imageFileNameDic.keys():
    if name not in labelFileNameDic:
        print(name)
        #os.remove(img_dest_dir + '/' + name + '.png')
        shutil.move(img_dest_dir + '/' + name + '.png', '/media/hooman/hsSsdPartUbuntu/FM_PROJECT/dataPreparation/FMDL_3.1/backhoe/fmdl-backhoe-trainingData-new_may10-2019/imagesWithNoLabels')
```

```python
# remove labels we deleted the images for

for name in labelFileNameDic.keys():
    if name not in imageFileNameDic:
        print(name)
        #os.remove(label_dest_dir + '/' + name + '.png')
        shutil.move(label_dest_dir + '/' + name + '.png', '/media/hooman/hsSsdPartUbuntu/FM_PROJECT/dataPreparation/FMDL_3.1/backhoe/fmdl-backhoe-trainingData-new_may26-2019/labelsWithNoImages')
```

# Converting jpg image to png, and removing the jpegs

```python
from PIL import Image

for name in imageFileNameDic.keys():
    print(name)
    im = Image.open(img_dest_dir + '\\' + name + '.jpg')
    im.save(img_dest_dir + '\\' + name + '.png')
    os.remove(img_dest_dir + '\\' + name + '.jpg')
```
